```python
#Bubble sort
def bubble_Sort(arr):
  n = len(arr)
  for i in range(n):
    for j in range(0,n-i-1):
      if arr[j] > arr[j+1]:
        arr[j],arr[j+1] = arr[j+1],arr[j]
  return arr

user_input  = input("Enter the elements separated by spaces ::")
my_list = list(map(int,user_input.split()))
print("Original Array: ", my_list)
sorted_list = bubble_Sort(my_list)
print("Sorted array : ",sorted_list)
```

```python
# Selection Sort

def Selection_Sort(arr,size):
  for i in range(size):
    min_index = i
    for j in range(i+1,size):
      if arr[j] < arr[min_index]:
        min_index = j
    (arr[i],arr[min_index]) = (arr[min_index],arr[i])

arr = [-2,34,34,5,54,77,-12,-120]
print(arr)
size = len(arr)
Selection_Sort(arr,size);
print(arr)
```

```python
#merge Sort
def Merge_Sort(arr,st,end):
  if st>=end:
    return
  mid = (st+end)//2
  Merge_Sort(arr,st,mid)
  Merge_Sort(arr,mid+1,end)
  Merge(arr,st,mid,end)

def Merge(arr,st,mid,end):
  temp = [0] * (end - st + 1)
  i = st
  j = mid+1
  k = 0
  while i<=mid and j<=end:
    if arr[i] < arr[j]:
      temp[k] = arr[i]
      i+=1
    else:
      temp[k] = arr[j]
      j+=1
    k+=1
  while i<=mid:
    temp[k] = arr[i]
    i+=1
    k+=1
  while j<=end:
    temp[k] = arr[j]
    j+=1
    k+=1
  for z in range(len(temp)):
        arr[st + z] = temp[z]


user_input = input("Enter the numbers separated by spaces:: ")
my_list = list(map(int, user_input.split()))
print("Original Array :: ", my_list)
length = len(my_list)

Merge_Sort(my_list, 0, length - 1)  # sort in place

print("Sorted List ::", my_list)  # print the modified list
```

```python
# Using Python to solve the following Transportation
# problems: A company distributes products to 3 retailers,
# (but total supply ‚â† total demand).
#           R1 R2 R3 Supply
# Source S1 10 15 12 40
# Source S2 18 14 16 50
# Demand    30 35 20


from pulp import *

# Define model (Maximization of profit)
model = LpProblem("Max_Profit", LpMinimize)

# Decision variables (row-wise: S = source, R = row)
x11 = LpVariable("S1_to_R1", lowBound=0)
x12 = LpVariable("S1_to_R2", lowBound=0)
x13 = LpVariable("S1_to_R3", lowBound=0)
x14 = LpVariable("S1_to_R4",lowBound = 0)

x21 = LpVariable("S2_to_R1", lowBound=0)
x22 = LpVariable("S2_to_R2", lowBound=0)
x23 = LpVariable("S2_to_R3", lowBound=0)
x24 = LpVariable("S2_to_R4", lowBound=0)



# Objective function: maximize profit
model +=10*x11 + 15*x12 + 12*x13+ 0*x14 + 18*x21 + 14*x22 + 16*x23 + 0*x24,"Total Cost"


model += x11 + x12 + x13 +x14 == 40, "Supply_S1"
model += x21 + x22 +x23 +x24 == 50, "Supply_S2"


# Demand constraints (column-wise)
model += x11 + x21 == 30, "Demand_R1"
model += x12 + x22 == 35, "Demand_R2"
model += x13 + x23 == 20, "Demand_R3"
model += x14 + x24 == 5, "Demand_R4"
# print(model)
# Solve the model
model.solve()

# Print results
print("\n--- Transportation Problem (Maximization) ---")
print("Status:", LpStatus[model.status])
print(f"S1->R1: {x11.varValue}, S1->R2: {x12.varValue},S1->R3:{x13.varValue},S1->R4:{x14.varValue}")
print(f"S2->R1: {x21.varValue}, S2->R2: {x22.varValue},S2->R3:{x23.varValue},S2->R4:{x24.varValue}")

print("Maximum Profit =", value(model.objective))
```

```python
# Use Golden Section Method to minimize the function  ùëì(ùë•) = (ùë• ‚àí 2)2 over
# the interval [0, 5]. Perform the search with tolerance level ùúñ = 0.00001  (based
# on Golden Ratio). Write a Python program to implement the method and find the
# approximate minimizer and the corresponding minimum value of the function.

def golden_section_search(func, a, b, tol=1e-8):

    phi = (1 + 5 ** 0.5) / 2
    resphi = 2 - phi

    x1 = a + resphi * (b - a)
    x2 = b - resphi * (b - a)

    f1 = func(x1)
    f2 = func(x2)

    while abs(b - a) > tol:
        if f1 < f2:
            b = x2
            x2 = x1
            f2 = f1
            x1 = a + resphi * (b - a)
            f1 = func(x1)
        else:
            a = x1
            x1 = x2
            f1 = f2
            x2 = b - resphi * (b - a)
            f2 = func(x2)

    return (a + b) / 2

f = lambda x: (x-2)**2
result_golden = golden_section_search(f, 0, 5)
min_val_golden = f(result_golden)
print("Approximate minimum at x =", result_golden)
print("Approximate minimum value is:", min_val_golden)
```
